# Problem.
Diberikan bilangan bulat positif $n$. Buatlah program Python untuk menghitung banyaknya pasangan bilangan bulat positif $(x,y)$ yang memenuhi persamaan $\displaystyle \frac{1}{x}+\frac{1}{y}=\frac{1}{n}$ dan menampilkan semua pasangan tersebut dalam sebuah tabel.

# Solusi:
Perhatikan bahwa
\begin{align*}
\frac{1}{x} + \frac{1}{y} = \frac{1}{n}
& \iff \frac{1}{x} = \frac{1}{n} - \frac{1}{y} \\
& \iff \frac{1}{x} = \frac{y-n}{ny} \\
& \iff x = \frac{ny}{y-n}.
\end{align*}
Karena $x$ merupakan bilangan bulat positif, haruslah $(y-n) \mid ny$ dan $y>n$.

Andaikan $y \le n$, diperoleh $\displaystyle \frac{1}{y} \ge \frac{1}{n}$ sehingga
$$\frac{1}{n}=\frac{1}{x}+\frac{1}{y} \ge \frac{1}{x}+\frac{1}{n} \implies \frac{1}{x} \le 0,$$
yang jelas tidak mungkin karena $x$ adalah bilangan asli.
Akibatnya, $n+1 \le y$.

# Implementasi Program:

```python
import numpy as np
import pandas as pd
from pandas import DataFrame

#Definisikan fungsi.
def jumlahresiprokal(n):
  if (n<0 or n==0):
    print(f'{n} bukan merupakan bilangan bulat positif!')
  else:
    x1 = []
    y1 = []
    #Kasus x>=y.
    for y in range(n+1,2*n+1):
      if ((n*y) % (y-n) == 0):
        x = (n*y) // (y-n) #(Floor division) Hasil bagi.
        x1.append(x)
        y1.append(y)
    #Kasus x<y
    for x in range(n+1, 2*n):
      if ((n*x) % (x-n) == 0):
        y = (n*x) // (x-n) #(Floor division) Hasil bagi.
        x1.append(x)
        y1.append(y)
    data={'x':x1, 'y':y1}
    df=pd.DataFrame(data)
    df.index = np.arange(1, len(df)+1)
    print(f'Banyaknya pasangan bilangan bulat positif (x,y) yang memenuhi 1/x + 1/y = 1/{n} adalah {len(df)}.')
    print(f'Semua pasangan bilangan bulat positif (x,y) yang memenuhi 1/x + 1/y = 1/{n} diberikan pada tabel berikut.')
    return(df)

n = int(input('Masukkan nilai n: '))
solusi=jumlahresiprokal(n)
solusi
```
